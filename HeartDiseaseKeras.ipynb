```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score #to evaluating my model performanc
```

```python
data = pd.read_csv("/heart.csv")
data.head()
data.tail()
```

```python
data.columns
```

```python
data.describe()
```

```python
data['target'].value_counts()
```

1 : Defective Heart
0 : Healthy Heart

```python
# taking dependent and idependent couln target take from x axis
x = data.drop(columns='target', axis=1) #independent features
x
```

```python
y = data['target']
y
```

Splitting test and  train data


```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2, stratify=y, random_state=2) #stratify is used to maintain the ratio of target variable in train and test data
print(x_train.shape,x_test.shape)
```

Model Training


```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import accuracy_score
```

```python
model = Sequential()
model.add(Dense(1, activation='sigmoid', input_shape=(x_train.shape[1],)))

model.compile(optimizer=Adam(0.01),
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
history = model.fit(x_train, y_train,
                    validation_data=(x_test, y_test),
                    epochs=50,
                    batch_size=32,
                    verbose=1)
```

Training with data

```python
train_pred = model.predict(x_train)
train_pred = (train_pred > 0.5).astype(int)
train_acc = accuracy_score(y_train, train_pred)

# Test accuracy
test_pred = model.predict(x_test)
test_pred = (test_pred > 0.5).astype(int)
test_acc = accuracy_score(y_test, test_pred)

print("Training Accuracy:", train_acc)
print("Test Accuracy:", test_acc)
```
